```python
# implementation of stack

class myStack:
    def __init__(self, cap):
        # array to store element 
        self.arr = [0]*cap

        # Maximum size of the stack
        self.capacity = cap

        # index of top element 
        self.top = -1

    def push_operation(self, x):
        if self.top == self.capacity - 1:
            print("Stack Overflow")
            return
    
        self.top += 1
        self.arr[self.top]=x
        return self.arr


obj1 = myStack(5)
obj1.push_operation(20)
obj1.push_operation(30)
obj1.push_operation(40)
obj1.push_operation(50)
        
```

# Operation on the stack

### Push Operation 

```python
def push_operation(self, x):
    if self.top == self.capacity - 1:
        print("Stack Overflow")
        return

    self.top += 1
    self.arr[self.top]=x

    
```

### Pop Operation 

```python
def pop(self):
    if self.top == -1:
        print("stack Underflow")
        return -1

    value = self.arr[self.top]
    self.top -= 1
    return value 
    
```

### Top or Peek Operation in Stack:

```python
def peek(self):
  
    if self.top == -1:
        print("Stack is Empty")
        return -1
  
    return self.arr[self.top]
```

### isEmpty Operation

```python
def isEmpty(self):
   
    return self.top == -1
```

### isFull Operation

```python
def isFull(self):
    return self.top == self.capacity - 1
```
